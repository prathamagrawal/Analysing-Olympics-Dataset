```python
import pandas as pd
import numpy as np
import seaborn as sn
import pandas_profiling
from  sklearn.preprocessing import  LabelEncoder
import matplotlib.pyplot as plt
```

```python
athletes=pd.read_csv("tokyo/athletes.csv")
medal=pd.read_csv("tokyo/medals.csv")
```

```python
print(athletes.shape)
print(medal.shape)
```

```python
athletes=athletes.drop(["Date of Birth","Age"],axis=1)
```

```python
athletes.head()
```

```python
medal.head()
```

```python
data=pd.DataFrame()
```

```python
data['Name']=athletes['Name']
allnames=list(data['Name'])
data['Sex']=athletes['Gender']
data['Age']=athletes['Age.1']
data['Team']=athletes['Country']
data["Games"]="2020 Summer"
data["Year"]="2020"
data['Season']="Summer"
data['Sport']=athletes['Sport']
data['Medal']=0
```

```python
data.head(10)
```

```python
names=list(medal['Name'])
medals=list(medal['Medal'])
l=[]
for i in allnames:
    if (i in names):
        index=names.index(i)
        value=medals[index]
        l.append(value)
    else:
        l.append(np.NaN)
data['Medal']=l
```

```python
data.head()
```

```python
data['Medal'].fillna(value=0, inplace=True)
data['Medal']=data['Medal'].replace('Gold Medal',3)
data['Medal']=data['Medal'].replace('Silver Medal',2)
data['Medal']=data['Medal'].replace('Bronze Medal',1)
```

```python
data.to_csv("2020_tokyo_olympics.csv")
```
